pengolahan data transaksi sederhana

```python
transaksi = [
    "U001:Buku:50000",
    "U002:Pensil:5000",
    "U001:Pulpen:10000",
    "U003:Penghapus:2000",
    "U002:Buku:50000"
]

def summary_transaksi(y):
  results = {}

  for item in y:
    parts = item.split(":") # pakai split untuk memotong teks berdasarkan karakter tertentu (misalnya titik dua :) dan mengubahnya menjadi List.
    user_id = parts[0]
    harga = int(parts[2]) # remember to change it into integer

    if user_id in results:
      results[user_id] += harga
    else:
      results[user_id] = harga


  return results

summary_transaksi(transaksi)

```

data cleaning and agregation

```python
import pandas as pd
import numpy as np

data = {
    'Siswa': ['Ali', 'Budi', 'Citra', 'Deni', 'Eka', 'Fani'],
    'Kelas': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Nilai': [80, np.nan, 90, 70, 85, np.nan]
}
df = pd.DataFrame(data)

def proses_nilai(df):
  rata_rata = df['Nilai'].mean() #df['Nilai'] --> kolom nilai

  df['Nilai'] = df['Nilai'].fillna(rata_rata)

#groupby : mengelompokkan data berdasarkan kategori tertentu
  results = df.groupby('Kelas')['Nilai'].mean() #mengelompokkan berdasarkan kelas lalu ambil kolom nilai, lalu cari rata2nya
  return results

proses_nilai(df)
```

```python
import pandas as pd
import numpy as np

data = {
    'Produk': ['Laptop', 'Mouse', 'Laptop', 'Keyboard', 'Mouse'],
    'Gudang': ['A', 'A', 'B', 'A', 'B'],
    'Stok': [10, 50, np.nan, 20, np.nan] # Ada NaN!
}
df = pd.DataFrame(data)

def stok(df):
  df['Stok'] = df['Stok'].fillna(0)

  result = df.groupby('Produk')['Stok'].sum()
  return result

stok(df)
```

calculate Euclidean Distance

```python
import math

def hitung_jarak(a,b):
  return math.sqrt(sum((x-y)**2 for x,y in zip (a,b)))

hitung_jarak([1,2,3],[4,6,8])
```

```python
import math

def jumlah_selisih_mutlak(a,b):
  return sum(abs(x - y) for x,y in zip(a,b))

jumlah_selisih_mutlak([1,2,3],[4,5,6])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('shopbehave.csv')
print(df)
#cek data kosong
#print(df.isnull().sum())

rata_belanja = df.groupby('Gender')['Purchase Amount (USD)'].mean()
print(f"\naverage spending by gender : \n{rata_belanja}")

df['Code'] = np.where(df['Promo Code Used'] == 'Yes',1,0)

penggunaan_promo = df.groupby('Gender')['Code'].sum()
print(f"\nstatistic people who use promo code \n{penggunaan_promo}")


df['transaction'] = np.where(df['Gender'] == 'Male',1,1)
gender = df.groupby('Gender')['transaction'].sum()
print(f"\nwho buy more : \n{gender}")

payment_favorite = df['Payment Method'].mode()[0]
print(f"\nthe favorite payment method is {payment_favorite}\n")

b = df.groupby('Gender')['Purchase Amount (USD)'].agg(['min','max','mean'])
print(b)

# index = Baris (mau dikelompokkan berdasarkan apa?)
# columns = Kolom (data apa yang mau dijadikan judul kolom?) -> Di sini 'Gender'
# values = Nilai (angka apa yang mau dihitung?)
# aggfunc = Rumus (mean/sum/count)


tabel_pivot = df.pivot_table(index='Category',
                             columns='Gender',
                             values='transaction',
                             aggfunc='sum')

print(f"\n{tabel_pivot}")
```
